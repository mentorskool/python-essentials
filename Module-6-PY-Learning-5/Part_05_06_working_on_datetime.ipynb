```python
df['diff_days']= df['starttime'] - df['stoptime']
```

**'starttime' and 'stoptime' column are in object format. For any type of operation on date time we have to set datatype as datetime**

```python
df['new starttime'] = pd.to_datetime(df['starttime'],format='%d-%m-%Y %H:%M')
df.head()
```

```python
df['new stoptime'] = pd.to_datetime(df['stoptime'],format='%d-%m-%Y %H:%M')
df.head()
```

```python
df.info()
```

```python
df['diff_days']= df['new starttime'] - df['new stoptime']
df.head()
```

```python
#for extracting only days as the output 
df['new diff_days'] = df['diff_days'].dt.days
df.head()
```

```python
#for extracting only date from the datetime as the output 
df['starttime_date'] = df['new starttime'].dt.date
```

```python
#for extracting only week number from the datetime as the output 
df['starttime_week'] = df['new starttime'].dt.isocalendar().week
```

```python
df.head()
```

```python
#similarly we can extract month year and also can present them in different different format
df['starttime_year'] = df['new starttime'].apply(lambda x: x.year)
df['starttime_month'] = df['new starttime'].apply(lambda x: x.month)
df['starttime_month_name'] =df['new starttime'].apply(lambda x: x.strftime('%b'))
df['starttime_year_month'] = df['new starttime'].apply(lambda x: x.strftime('%Y%m'))
df['starttime_day'] =df['new starttime'].apply(lambda x: x.day)
df['starttime_day_of_week'] =df['new starttime'].apply(lambda x: x.dayofweek)
df['starttime_dow_name'] =df['new starttime'].apply(lambda x: x.strftime('%a'))
```
