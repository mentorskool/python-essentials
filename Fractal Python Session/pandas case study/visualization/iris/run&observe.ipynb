```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('Iris.csv')
iris_feat = iris.iloc[:,:-1]
iris_species = iris.iloc[:,-1]
```

# Scatter Plot

* It is best suited for situations where the dependent variable can have multiple values for the independent variable.

### sepal_length vs petal_length

```python
plt.scatter(iris_feat['SepalLengthCm'],iris_feat['PetalLengthCm'],alpha=0.5)  # alpha changes the transparency
# set a title and labels
plt.title('Scatter Plot')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
#Show the plot
plt.show()
```

# Scatter plot for Multivariate Analysis

### 'sepal_length' VS 'sepal_width'
*  Use color accordingly
    * Iris-setosa: red
    * Iris-virginica : green
    * Iris-versicolor : blue

```python
colors = {'Iris-setosa':'r', 'Iris-virginica':'g', 'Iris-versicolor':'b'}

# create a figure and axis
fig, ax = plt.subplots()

# plot each data-point
for i in range(len(iris_feat['SepalLengthCm'])):
    ax.scatter(iris_feat['SepalLengthCm'][i], iris_feat['SepalWidthCm'][i],color=colors[iris_species[i]])

# set a title and labels    
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')

plt.show()
```

# Line Plot

* Line plots can help display data points for both single variable analysis as well as multiple variable analysis.

```python
# get columns to plot
columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
# create x data
x_data = range(0, iris.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, iris[column], label=column)
# set title and legend
ax.set_title('Iris Dataset')
ax.legend()
```

# Histograms

```python
fig, ax = plt.subplots()
# plot histogram
ax.hist(iris_feat['SepalLengthCm'])
# set title and labels
ax.set_title('sepal_length')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

# Bar Chart

```python
df = iris.groupby('Species')['SepalLengthCm'].sum().to_frame().reset_index()
#Creating the bar chart 
plt.bar(df['Species'],df['SepalLengthCm'],color = ['cornflowerblue','lightseagreen','steelblue']) 
#Adding the aesthetics
plt.title('Bar Chart')
plt.xlabel('Species')
plt.ylabel('sepal_length') 
#Show the plot
plt.show()
```
