```python
# Let's move on to the next scenario : 
# Which customer's name is the longest?

# Let's read the customer names in a variable
names = ['Tamara Dahlen','Charles Crestani','barry blumstein','Ted Braxton','Tom Christensen']
max_len = 0
name_max_len = ''

# Step 1 : Start looping
for name in names :

# fetch the length of the name. Just like len() function helps find the length of a list, you can use it to find length of a string too!
  length = len(name)

# Check if the length is greater than or equal to max_len
# Since max_len is = 0, new value of max_len will be the length of the customer name Tamara Dahlen in first pass
# The variable name_max_len will store the name of the first customer Tamara Dahlen in the first pass
  if length >= max_len :
    max_len = length
    name_max_len = name

# From next pass onwards, if the length of the next customer's name is greater than first customer, max_len and name_max_len variables will be replaced with the next customer's name
# Finally, once out of loop, print the values of variables name_max_len and max_len using formatted print statement

print(f'The Customer {name_max_len} has the longest name of size {max_len}',)

```

```python
# Let's look at another way to achieve the same result
# This one uses max() and index() functions along with len()

# Create an empty list. We will add length of all names here
len_name = []

# Start looping through the list names
for i in names:
  # create a variable x where store length of the name
  x = len(i)
  #add the length(numeric value) into the new list 
  len_name.append(x)

# let's check how the list look like
print(len_name)

# using the max function we can find out max numeric value of the list
max_len = max(len_name)
print(max_len)   

# as we access a value of list through index, similarly using the index function we can fetch the index of a particular value of the list
max_len_index = len_name.index(max_len)
print(max_len_index)

# as we put the length of the names in the new list according to the names 
# it's obvious the index of max len in new list will represent the longest name in the name list
print("The longest customer's name : ", names[max_len_index])
```

```python
# Next Scenario : 
# Customer name Rob Lucas needs to be split into two parts, where first name should be Rob and last name should be Lucas

name = 'Rob Lucas'

# Create two empty string. We will add length of all names here
first_name = ''
last_name = ''
# as using the index function we can fetch the index of a particular value of the list, similarly we can do that with string 
space_index=name.index(' ')
print(space_index)

# Start looping through the character of strings within the range (starting to before ' ')
for i in range(0,space_index):
  first_name=first_name + name[i]

# Start looping through the character of strings within the range (After ' ' to end)
for i in range(space_index,len(name)):
  last_name=last_name + name[i]

print('first name:',first_name)
print('last name:',last_name)
```

```python
# Next Scenario :
# Suppose Customer's name is :Rob Ivan Lucas
# How will you split the name into three parts : 'Rob', 'Ivan' and 'Lucas'

name='Rob Ivan Lucas'

split_value = []
tmp = ''

for c in name:
    if c == ' ':
        split_value.append(tmp)
        tmp = ''
    else:
        tmp += c # this is same as tmp = tmp + c
if tmp:
    split_value.append(tmp)

print(f'First Name : {split_value[0]}')
print(f'Middle Name : {split_value[1]}')
print(f'Last Name : {split_value[2]}')
```

```python
# Did the above code work?
# Did you copy paste the error on Google and explore why?
# Well, the reason is :
# List as a data type is mutable which means it can be edited
# String as data type is immutable which means it can't be edited. 
```

```python
# Example for list : 
list_example = ['b','a','r','r','y']
list_example[0] = 'h'
list_example

# Here the list got successfully edited to harry. That's what mutable means
```

```python
# Example for string : 
string_example = 'barry'
string_example[0] = 'h'
string_example

# Here string couldn't be edited. That's what immutable means
```

```python
string_example = 'barry'

# We use replace method to change b to h but in this case the result is alloted back the variable string_example
# Thus the full string got replaced by new string

string_example = string_example.replace('b','h')
string_example
```
