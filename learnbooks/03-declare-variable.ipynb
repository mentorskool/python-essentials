**Task - 1 :**

Let’s look at the first record in the **Orders** sheet. The **Sales Amount** for the product **Bush Somerset Collection Bookcase** purchased by customer_id **AA-10315** is **2000**. The total **quantity** sold is **2**. 

How would you calculate the **unit price** in Excel?


You’ll write a formula referring the **cell addresses** of **2000** and **2** and the outcome would look something like below : 

![](https://raw.githubusercontent.com/mentorskool/python-essentials/main/Module-6-PY-Learning-4/imgs/var_intro_sample_img1.png)


What’s the **cell address** here? It is nothing but a way for Excel or rather your computer to **memorize the values** **2000** and **2**. Also, you can change the values of the cell addresses **K2** and **L2** to anything you want.  In short, Its value **varies**. Hence, it is called a **variable**

How would you achieve the above task in **Python**?

```python
# First, let's try doing this in the simplest way. Divide 2000 by 2 
# In order to see the output execute this cell by clicking on the play button or hit Shift + Enter

2000/2
```

```python
# Next, let’s see how to fetch the same output by using variables - 
# Store the values in two variable named sales and quantity

sales = 2000
quantity = 2

# Divide sales by quantity and store in a new variable called output

output = sales/quantity

# Let's see what does the variable output stores

output 
```

```python
# Also you can use print and put the variable inside round brackets, remove'#' and run the code. 
# This '#' symbol is nothing but a way to ensure the code following this gets ignored by Python compiler. 
# Use it freely to add comments to your code as we have done here

# print(output)
```

**Task - 3 :**

Given the **sales amount** of **2000**, **quantity** of **2** and **profit** of **500** (Check snapshot of **order details** above), What will be the **unit cost** at which the product was sold? (**Ignore discount** for this task)

```python
# Let's store profit value in a new variable

profit = 500

# Think what will you do in Excel to calculate the unit cost of product 
# We will do something similar in Python

output = sales-profit/quantity
print(output)
```

**Note :** 

* Did you observe that we **didn't need to declare** sales and quantity variable again in the above cell because it **picked up the value** from the variables already **declared** and **assigned** earlier. (This helps ensure values are **assigned at a single place** and used wherever needed)

* Also, observe that the value of **output** variable which was **1000.0** earlier got replaced by **1750.0**. This shows that **variables can be overwritten** by new operations. If you wish to retain value of earlier variable, try giving a **different variable name** like **output1** or something
