As a part of their **end of season** sales, **GlobalMart** wishes to offer **flat 20% discount** on **sales amount** of all **item**s belonging to **Furniture** category

* How can we **store** them?
* Once stored, how can we **process** them?

* Enter **sequence data types**, specifically **Lists**

```python
# Let's store only first 2 values for now
# Lists are created using square bracket
sales_amounts = [2000,700]
print(sales_amounts)

# Let's check the data type of the variable sales_amounts
print(type(sales_amounts))
```

* Did you **observe** a **pattern** above?
* We are **accessing items** of a **list** by using a **combination** of **square brackets [ ]** and a **number** starting from **0** and followed by **1,2,3 etc**.
* This idea of **using numbers** to **access items** in a list is the idea of **indexing** 

Check the **image** below to **understand** how **values of a list** are related to the **index(position)** of the values :      

![](https://msklbusinessdata.blob.core.windows.net/python-masterclass/06-indexing.png)

**Important Note :** 
* The **index** (or **position**) of **items** in a list **starts with 0 rather than 1** (which comes **intuitively** in mind). This may seem **wierd** but that's how programmers are. Let's go by **what's the rule**. 
* This would mean in a list of **2 items**, the **last value** will have an **index 1** rather than **2**

**Scenario - 4 :**

Now, let's go back to **Scenario - 1** where we were trying to **print** the **discount amounts**. 

We tried the below code :       

![](https://msklbusinessdata.blob.core.windows.net/python-masterclass/06-lists-01.png)

* Do you see a **problem** here?
  * We have to **declare** a **new variable** for **discount calculation** for **every item member** of the list **again and again**
  * The code is getting **repeated**
  * Imagine doing it for a list with **100 items!**

* How do we **solve** this problem?
* How do we **solve** a **general problem** of conveniently **repeating a set of tasks** without writing a **lot of repeated code**?

* **Enter** fundamental programming concept of **looping**!

```python
# Let's check the below code :

for x in sales_amounts :
  print('The discount amount is',x*0.2)
```

**Wait**, What just happened?

**2 lines of code** and we are **done!!**

That's the **power of looping** and **simplicity** of doing that in **Python**!

Let's **read the code** this way :    

**For** every **item** belonging to the list **sales_amounts**, **store** the item in **variable x**, then **multiply x by 0.2 (20%)** and keep **printing the results** until we reach the **end of the list**

**Scenario - 5 :**

Refer to the **Sales** and **Quantity** column of the **order details** once again :

![](https://msklbusinessdata.blob.core.windows.net/python-masterclass/03-order-details-03.png)

How do we **print** the **unit prices** using **loops**?

Well, to **calculate** unit prices, we need **access** to both **Sales** and **Quantity** values **at once**.

Let's **try** this below : 

```python
# Let's store sales values in one list
sales = [2000,700,900,50,500,200]

# Let's store quantity values in another list
qty = [2,3,5,7,6,9]

# Now check out the below cod
for i in range(len(sales)):
  print('The unit price is :',sales[i] / qty[i])
```

Let's **dissect** the **code** above :      

* We already know the role of **for**
* **i** here is a **variable** as described before. The **difference** here is that instead of storing the **actual item** of the list, it stores the **index of that item**
* **len()** is a **function** which returns the **length** (**total count of members**) of the list
* **range(x)** is a function which **generates** values from **0** to **one less than number x**. This is **needed** because the **variable i** needs to **loop** through numbers from **0 to x-1** instead of **1 to x**

Continuing with **Scenario - 5**, How can we **write** the **unit price** in a **better way** as shown below?

* The **unit price** of **Bush Somerset Collection Bookcase** is **1000.0**

```python
# Let's store sales values in one list
sales = [2000,700,900,50,500,200]

# Let's store quantity values in another list
qty = [2,3,5,7,6,9]

# Let's store product names in another list
products = ['Bush Somerset Collection Bookcase',
'Hon Deluxe Fabric Upholstered Stacking Chairs',
'Bretford CR4500 Series Slim Rectangular Table',
'Eldon Expressions Wood and Plastic Desk Accessories',
'Mitel 5320 IP Phone VoIP phone',
'Chromcraft Rectangular Conference Tables']

# Now check out the below cod
for i in range(len(sales)):
  print('The unit price of ' + products[i] + ' is : ',round(sales[i] / qty[i],2)) # Recall operator overloading? 
  
# In the above line, three pieces of texts are being concatenated together with the help of + symbol
# Part 1 : 'The unit price of '
# Part 2 : product[i] which is the name of the product at ith index of the list
# Part 3 : ' is : '
# Also, check the usage of round() function to round off the excess digits
```

**Important Follow up discussion :**

We observed how we can write more **nicely readable print statements** above. Let's look at **few more cases** and try printing well formatted output. 

* Given Sales amount as **1000** and discount value as **20.5%**, How will you print a statement like :  **The discount applied on amount of 1000.0 is 20.5%**

* Given Customer name as **Jack Ryan**, date of birth as **22-Feb-1992** and credit score as **9.8**, How will you print a statement like : **Customer Jack Ryan was born on 22-Feb-1992 and has a credit score of 9.8**

[Here](https://realpython.com/python-f-strings/) is a **good read** on the **journey** of **print** statement's **output formatting**. 
It started with **%formatting**, **str.format()** and evolved into an **easy** and **convenient** **f-string** method. Do spend some time!

```python
# Examples of % literals
sales_amount = 1000
discount = 20.5
customer_name = 'Jack Ryan'
dob = '22-Feb-1992'
credit_score = 9.8

print("The discount applied on amount of %d is %.1f" %(sales_amount, discount))
print("Customer %s was born on %s and has a credit score of %.1f" %(customer_name,dob, credit_score))

```

```python
# Examples of f-literals
sales_amount = 1000
discount = 20.5
customer_name = 'Jack Ryan'
dob = '22-Feb-1992'
credit_score = 9.8

print(f"The discount applied on amount of {sales_amount} is {discount}")
print(f"Customer {customer_name} was born on {dob} and has a credit score of {credit_score}")

# Undoubtedly, f-literals are super easy and win hands down. No wonder, Python developers have adopted f-literals to perform formatted printing. %literals are old school now!
```
