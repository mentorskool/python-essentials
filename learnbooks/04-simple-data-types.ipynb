* What we did **above** is an **example** of **Type Conversion** task in **Python**
* **Numbers** in Python can **easily be converted** from a **given type** to a **target type**. Each numeric type has an associated **conversion function** which accepts a number and converts it to the desired type. For example a number may be converted to an int with the **int()** function, to a float with the **float()** function, and to a complex with the **complex()** function
* Also we can use** round function** if we have to trim digits after decimal to some fixed length. Check the code below

```python
# Let's convert 233.33333333333334 (the output of 700/3) fixed upto 2 decimal 233.33

round(700/3,2)

# Here, 2 decides how many digits to be kept after the decimal
```

* **Small Note :** In **type conversion**, we just **change** the **output format**. The **actual valu**e not converted

```python
# Let's observe this through below code. Also, learn how we can print nicely worded output using print function

x=700/3
print('the original value of x',x) 
print('type of x',type(x))
print('convert value of x in integer',int(x))
print('round up the value of x upto 2 decimal',round(x,2))
print('Again print the original value of x',x)
```

* What if you wish to **save the output** with **changed data type** for further usage?
* Simple, **store the output** of type conversion to a **different variable**



```python
# Try the below code
x=700/3
print('the original value of x',x)
y=int(x) #declare a new variable y and store the int value of x
print('the value of y',y)
print('the value of x after declaring y',x)
x=round(x,2) # declare the same variable x and store the value. Here we are loosing the original value of x.
print('the value of x after storing the new value in x',x)
```
