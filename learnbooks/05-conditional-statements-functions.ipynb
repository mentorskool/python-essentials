```python
# As we now the benifit of variable,let's declare
sales = 2000
ship_mode = 'First class'

# you are already familiar with if else condition ,just look at the format 
# if the condition :
#   value if true
# else:
#   value if false

if ship_mode == 'First class':
  x = sales+sales*(10/100)      #benefit in python, you can put more complex calculation here, not only the single value
  print(x)                    
else:
  x=sales+sales*(5/100)
  print(x)
```

**Scenario - 3 :**

If you can **observe** in **Scenario - 2**, we did the **calculation** for **just one Order**. How will you implement the same for the **2nd order**, **3rd order** etc.?

**Simple**, **repeat** the same code **again** and **again**. 

* However, don’t you think that’s a bit **boring** and leads to **repeated code** at **so many places**?

* Also, think tomorrow if the **surchage rates** are **modified**, how **many places** will you have to **fix your code** to achieve **correct results**?

* What's the **solution** to this?

What if we could write the steps to perform this task once and for all, and call that task wherever needed?

That’s where **the idea of functions** come into picture.

**Explore** more about **Functions** [here](https://www.w3schools.com/python/python_functions.asp)

Let's check the **below cod**e to **understand** how to **write a function** :

```python
# define a function def keyword 
# write the name of function
# followed by round bracket inside which we pass the variables needed for the logic inside the function
# then use return keyword to return your output of the function
def surcharge(ship_mode,sales): 
  if ship_mode=='Same Day':
    x = sales+sales*(20/100)                 
  elif ship_mode=='First class': 
    x = sales+sales*(15/100)
  elif ship_mode=='Standard class':
    x = sales+sales*(5/100)     
  else:
    x = sales
  return x

# Observe what we did above
# We simply encapsulated the entire conditional logic we wrote before within a function
# How does that help us fix problems mentioned earlier?
```

```python
# Type the above code yourself for practice

```

```python
# Let's call the function independently now
# For 1st Order
sales = 2000 
ship_mode = 'First class'

# call the function now
surcharge(ship_mode,sales)

# For 2nd Order
sales = 700
ship_mode = 'Standard class'
surcharge(ship_mode,sales)
```
